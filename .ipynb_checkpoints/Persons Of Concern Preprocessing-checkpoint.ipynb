```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
person_of_concern_data = "./refugee-data/persons_of_concern.csv"
data = pd.read_csv(person_of_concern_data)
```

```python
cols = ['Refugees (incl. refugee-like situations)', 'Asylum-seekers (pending cases)', 'Returned refugees', 'Stateless persons',
       'Others of concern', 'Total Population']

data[cols] = data[cols].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
data_post_year_2001 = data[data['Year'] > 2000]
```

```python
data_post_year_2001 = data_post_year_2001.fillna(0)
```

```python
data_post_year_2001.replace(to_replace ="*", value ="0")
```

```python
df = data_post_year_2001.reset_index(drop=True)
df.head()
```

```python
df.to_csv(r'refugee-data/Cleaned_people_of_concern.csv')
```
