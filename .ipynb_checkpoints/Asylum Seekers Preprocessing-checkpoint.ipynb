```python
import pandas as pd
data = pd.read_csv("refugee-data/asylum_seekers.csv")
```

```python
cols = ['Tota pending start-year', 'of which UNHCR-assisted(start-year)', 'Applied during year', 'decisions_recognized', 'decisions_other', 'Rejected', 'Otherwise closed', 'Total decisions', 'Total pending end-year', 'of which UNHCR-assisted(end-year)']
data[cols] = data[cols].apply(pd.to_numeric, errors='coerce', axis=1)
data
 
```

```python
data = data[data.Year != 2000] # removing all the rows with year below 2001
print(data['Year'].value_counts(dropna=False))
```

```python
data = data.fillna( value = 0)
```

```python
data.isnull().sum()
```

```python
data.replace(to_replace ="*", value ="0")
```

```python
data.to_csv("refugee-data/cleaned_asylum_seekers.csv")
```
