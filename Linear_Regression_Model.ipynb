```python
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import statsmodels.api as sm
```

```python
# Load the dataset
data = pd.read_csv("cleaned_data/cleaned_asylum_seekers_added.csv")
data = data[data['HDI'] != '..']
data = data.dropna()
decimals = 2    
data['Unemployment rate'] = data['Unemployment rate'].apply(lambda x: round(x, decimals))
data['origin_to_target_dist'] = data['origin_to_target_dist'].apply(lambda x: round(x, decimals))
 
data.head()
```

```python
col_names = [ 'Unemployment rate','HDI'] # Declare the columns names

x = data[col_names] 
y = data['Total decisions']



model = sm.OLS(y.astype(float), x.astype(float)).fit()
predictions = model.predict(x)

model.summary()
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, y, test_size=0.2)
print (x_train.shape, y_train.shape)
print (x_test.shape, y_test.shape)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(x_train, y_train)
predictions = lm.predict(x_test)
```
