```python
def lat(x):
    a = geocoder.arcgis(x)
    if(a.json is None):
        return ''
    else:
        return a.json['lat']
def long(x):
    a = geocoder.arcgis(x)
    if(a.json is None):
        return ''
    else:
        return a.json['lng']

# demo['lat'] = demo['locname'].apply(lat)
# demo['long'] = demo['locname'].apply(long)
```
