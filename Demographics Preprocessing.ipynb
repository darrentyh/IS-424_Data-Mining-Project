```python
import pandas as pd
import re
import geotext
import geocoder
```

```python
#read CSV file 
demo = pd.read_csv('refugee-data/demographics.csv')
```

```python
# get DataFrame info and create new headers
demo.info()
new_headers = ['year', 'country', 'locname', '0-4f', '5-11f', '5-17f', '12-17f', '18-59f', '60+f', 'unkf', 'totalf', '0-4m', 
               '5-11m', '5-17m','12-17m','18-59m', '60+m', 'unkm', 'totalm','lat','long']
```

```python
#dropping the index column
demo = demo.drop(columns = 'Unnamed: 0')
```

```python
#checking datatypes
demo.dtypes
```

```python
# replacing empty location names with ''
demo['locname'] = demo['locname'].fillna('')
```

```python
# replace all '*' with 0.0
demo = demo.replace(to_replace ="*", value =0.0) 
```

```python
#Fill NA with 0.0 and change datatype from string to integer
demo[['0-4f', '5-11f', '5-17f', '12-17f', '18-59f', '60+f', 'unkf', 'totalf']] = demo[['0-4f', '5-11f', '5-17f', '12-17f', '18-59f', '60+f', 'unkf', 'totalf']].fillna(0.0)
demo[['0-4m', '5-11m', '5-17m', '12-17m', '18-59m', '60+m', 'unkm', 'totalm']] = demo[['0-4m', '5-11m', '5-17m', '12-17m', '18-59m', '60+m', 'unkm', 'totalm']].fillna(0.0)
demo[['0-4f', '5-11f', '5-17f', '12-17f', '18-59f', '60+f', 'unkf', 'totalf']] = demo[['0-4f', '5-11f', '5-17f', '12-17f', '18-59f', '60+f', 'unkf', 'totalf']].astype('float64')
demo[['0-4m', '5-11m', '5-17m', '12-17m', '18-59m', '60+m', 'unkm', 'totalm']] = demo[['0-4m', '5-11m', '5-17m', '12-17m', '18-59m', '60+m', 'unkm', 'totalm']].astype('float64')
```

```python
# clean the location name
def check(x):
    a = re.sub('[^A-Za-z ]+','', x)
    return a

demo['locname'] = demo['locname'].apply(check)
```

```python
# functions to get the latitude and longitude of a location
def lat(x):
    a = geocoder.arcgis(x)
    if(a.json is None):
        return ''
    else:
        return a.json['lat']
def long(x):
    a = geocoder.arcgis(x)
    if(a.json is None):
        return ''
    else:
        return a.json['lng']
```

```python
# converting the location name to a string to iterate
a = demo['locname'].to_list()
lati = []
longi = []
```

```python
# adding to the lati and longi list the respective lats and longs of the location
for i in range(len(a)):
    lati.append(lat(a[i]))
    longi.append(long(a[i]))
```

```python
#adding two new columns
demo['lat'] = pd.DataFrame(lati)
demo['long'] = pd.DataFrame(longi)
```

```python
# changing new header name
demo.columns = new_headers
```

```python
# drop any NA data
demo = demo.dropna()
```

```python
# create new csv file for cleaned
demo.to_csv('cleaned_demographics.csv')
```
